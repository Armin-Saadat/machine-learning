```python
data = data.sample(frac=1).reset_index(drop=True)
data['bayas'] = 1


train = data.head(int(0.7 * 1436))
left = data.tail(1436 - int(0.7 * 1436))
validation = left.head(int(0.5 * 431))
test = left.tail(431 - int(0.5 * 431))


x_train = train.drop("Price", axis=1).to_numpy()
y_train = train["Price"].to_numpy()

x_validation = validation.drop("Price", axis=1).to_numpy()
y_validation = validation["Price"].to_numpy()

x_test = test.drop("Price", axis=1).to_numpy()
y_test = test["Price"].to_numpy()
```

<div dir="rtl">
    <h2>
          بخش اول
    </h2>
</div>
<p></p>
<div dir="rtl">
در این بخش از روش SGD با تابع خطا L2 استفاده میکنیم.
    بنابراین رابطه آپدیت کردن پارامترها به شکل زیر است:
</div>
$$
w^{i +1} = w^{i} + \eta(t_n - w^{(i)T} \phi_n)\phi_n
$$
<p></p>
<div dir="rtl">
        که eta همان rate است و رابطه سمت راست نیز مشتق L2 نسب به W است.
</div>

```python
def get_SSE(Y, X, W):
    return (np.sum(Y - X.dot(W))**2)

def update_w(W, batch_size, rate):    
    selected = np.random.choice(x_train.shape[0], batch_size, replace=False)
    x_batch = x_train[selected]
    y_batch = y_train[selected]
    return W + rate * (y_batch - x_batch.dot(W)).dot(x_batch)

def SGD_L2_simple(rate):
    W = np.zeros(x_test.shape[1])

    for i in range(2000):
        W = update_w(W, 25, rate)
    
    return get_SSE(y_validation, x_validation, W)

def SGD_L2(rate, iterations):
    validation_loss, test_loss = list(), list()
    W = np.zeros(x_test.shape[1])

    for i in range(iterations):
        W = update_w(W, 25, rate)
        validation_loss.append(get_SSE(y_validation, x_validation, W))
        test_loss.append(get_SSE(y_test, x_test, W))
    
    return validation_loss, test_loss, W
    
    
```

<div dir="rtl">
    در بخش بالا الگوریتم SGD پیاده‌سازی شد.
    <p></p>
        برای این پیاده‌سازی batch_size برابر با ۲۵ در نظر گرفت شد و تعداد دفعات تکرار نیز ۲۰۰۰ قرار داده شد.
    <p></p>
    در این پیاده‌سازی یک hyper_parameter در نظر گرفته شد که همان rate یادگیری است.
    <p></p>
    در ادامه با در نظر گرفتن چندین حالت برای rate حالت بهینه روی validation_set  به دست می‌آید.
</div>

```python
rates = np.arange(0.00001, 0.001, 0.00001)

best_validation_loss = 1000
best_rate = 0
for rate in rates:
    validation_loss = SGD_L2_simple(rate)
    if validation_loss < best_validation_loss:
        best_validation_loss = validation_loss
        best_rate = rate
        
best_validation_loss, best_rate
```

<div dir="rtl">
    بنابراین یک rate نسبتا بهینه به دست آمد.
    از آن برای سنجیدن خطای روی داده‌های تست و رسم نمودار استفاده میکنیم.
</div>

```python
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = [12, 8]
plt.rcParams['figure.dpi'] = 100 

validation_loss, test_loss, W_L2 = SGD_L2(best_rate, 2000)

plt.plot(validation_loss, label='validation_loss')
plt.plot(test_loss, label='test_loss')
plt.title("MSE Plot With Rate {} for {} iterations".format(round(best_rate, 6), 2000))
plt.legend()
plt.show()
print("MSE for test set is: " + str(test_loss[-1]))
```

<div dir="rtl">
    <h3>
        پیاده‌سازی ridge regressor از طریق SGD
    </h3>
</div>
<p></p>
<div dir="rtl">
       کاملا شبیه بخش ۱ عمل میکنیم با این تفاوت که در رابطه به روز رسانی پارامتر یک تفاوت کوچک وجود دارد و w قبلی نیز با w جدید جمع می‌شود.
</div>

```python
def get_SSE_ridge(Y, X, W, h):
    return (np.sum(Y - X.dot(W))**2) + h * np.sum(W.T.dot(W))

def update_w_ridge(W, batch_size, rate):    
    selected = np.random.choice(x_train.shape[0], batch_size, replace=False)
    x_batch = x_train[selected]
    y_batch = y_train[selected]
    return W + rate * ((y_batch - x_batch.dot(W)).dot(x_batch) - W)

def SGD_ridge(rate, iterations):
    validation_loss, test_loss = list(), list()
    W = np.zeros(x_test.shape[1])

    for i in range(iterations):
        W = update_w_ridge(W, 25, rate)
        validation_loss.append(get_SSE_ridge(y_validation, x_validation, W, 1))
        test_loss.append(get_SSE_ridge(y_test, x_test, W, 1))
    
    return validation_loss, test_loss, W
    
    
```

```python
validation_loss, test_loss, W_ridge = SGD_ridge(0.0001, 2000)

plt.plot(validation_loss, label='validation_loss')
plt.plot(test_loss, label='test_loss')
plt.title("MSE Plot With Rate {} for {} iterations".format(0.0001, 2000))
plt.legend()
plt.show()
print("MSE for test set is: " + str(test_loss[-1]))
```

<div dir="rtl">
    <h3>
    پیاده‌سازی با استفاده از فرمول بسته
    </h3>
</div>
<p></p>

<p></p>
<div dir="rtl">
    که در این تمرین phee همان X است.
    مشابه تمرین ۳ عمل میکنیم.
    درصمن لامبدا هم برابر با ۱ قرار میدهیم. مطابق صورت مسئله.
</div>

```python
phee = x_train
phee_t = phee.transpose()
phee_t_mul_phee = np.matmul(phee_t, phee)
inverse = np.linalg.pinv(phee_t_mul_phee + np.identity(x_train.shape[1]))
W_ridge = np.matmul(np.matmul(inverse, phee_t), y_train)
W_ridge
```

<div dir="rtl">
   با محاسبه moore penrose pseudo inverse مقدار بهینه برای پارامترها حساب شد.
<p></p>
بنابراین مقدار تارگت برای هر داده از ضرب w_transpose در بردار ویژگی‌های آن داده حساب می‌شود.    
</div>


```python
validation_loss_with_ridge = get_SSE_ridge(y_validation, x_validation, W_ridge, 1)
test_loss_with_ridge = get_SSE_ridge(y_test, x_test, W_ridge, 1)
print("MSE for validation set is: " + str(validation_loss_with_ridge))
print("MSE for test set is: " + str(test_loss_with_ridge))
```

<div dir="rtl">
    <h2>
          بخش سوم
    </h2>
</div>
<p></p>
<div dir="rtl">
در این بخش از روش SGD با تابع خطا L1 استفاده میکنیم.
</div>
$$
||Y - XW||_1
$$
<p></p>

```python
def get_L1_error(Y, X, W):
    return np.sum(np.abs(Y - X.dot(W)))

def update_w_L1(W, batch_size, rate):    
    selected = np.random.choice(x_train.shape[0], batch_size, replace=False)
    x_batch = x_train[selected]
    y_batch = y_train[selected]
    
    x_l1 = x_batch
    x_l1[(y_batch - x_batch.dot(W)) < 0] = -1 * x_l1[(y_batch - x_batch.dot(W)) < 0]
    
    return W + rate * np.sum(x_l1, axis=0)

def SGD_L1_simple(rate):
    W = np.zeros(x_test.shape[1])

    for i in range(2000):
        W = update_w_L1(W, 25, rate)
    
    return get_L1_error(y_validation, x_validation, W)

def SGD_L1(rate, iterations):
    validation_loss, test_loss = list(), list()
    W = np.zeros(x_test.shape[1])

    for i in range(iterations):
        W = update_w_L1(W, 25, rate)
        validation_loss.append(get_L1_error(y_validation, x_validation, W))
        test_loss.append(get_L1_error(y_test, x_test, W))
    
    return validation_loss, test_loss, W
```

<div dir="rtl">
    در بخش بالا الگوریتم SGD_L1 پیاده‌سازی شد.
    <p></p>
        برای این پیاده‌سازی batch_size برابر با ۲۵ در نظر گرفت شد و تعداد دفعات تکرار نیز ۲۰۰۰ قرار داده شد.
    <p></p>
    در این پیاده‌سازی یک hyper_parameter در نظر گرفته شد که همان rate یادگیری است.
    <p></p>
    در ادامه با در نظر گرفتن چندین حالت برای rate حالت بهینه روی validation_set  به دست می‌آید.
</div>

```python
rates = np.arange(0.00001, 0.001, 0.00001)

best_validation_loss = 1000
best_rate = 0
for rate in rates:
    validation_loss = SGD_L1_simple(rate)
    if validation_loss < best_validation_loss:
        best_validation_loss = validation_loss
        best_rate = rate
        
best_validation_loss, best_rate
```

<div dir="rtl">
    بنابراین یک rate نسبتا بهینه به دست آمد.
    از آن برای سنجیدن خطای روی داده‌های تست و رسم نمودار استفاده میکنیم.
</div>

```python
validation_loss, test_loss, w_temp = SGD_L1(best_rate, 2000)

plt.plot(validation_loss, label='validation_loss')
plt.plot(test_loss, label='test_loss')
plt.title("MSE Plot With Rate {} for {} iterations".format(round(best_rate, 6), 2000))
plt.legend()
plt.show()
print("MSE for test set is: " + str(test_loss[-1]))
```

```python
plt.plot(validation_loss[0:1000], label='validation_loss')
plt.plot(test_loss[0:1000], label='test_loss')
plt.title("MSE Plot With Rate {} for {} iterations".format(round(best_rate, 6), 1000))
plt.legend()
plt.show()
print("MSE for test set is: " + str(test_loss[999]))
```

<div dir="rtl">
    مشاهده می‌شود که واریانس میزان خطا در مقادیر کم زیاد است.
    بنابراین میتوان نرخ یادگیری کمتری را قرار داد.
</div>

```python
validation_loss, test_loss, W_L1 = SGD_L1(0.0001, 2000)

plt.plot(validation_loss, label='validation_loss')
plt.plot(test_loss, label='test_loss')
plt.title("MSE Plot With Rate {} for {} iterations".format(0.0001, 2000))
plt.legend()
plt.show()
print("MSE for test set is: " + str(test_loss[-1]))
```
