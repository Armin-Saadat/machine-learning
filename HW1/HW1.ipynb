```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

jobs = pd.read_csv("usjobs.csv")
jobs.head(5)
```

<div dir="rtl">
یه اتفاق بدی که معمولا برای دیتاست‌ها میوفته اینه که یک‌سری از داده‌ها تهی هستن و این کار معمولا تحلیل داده رو برای 
  ما سخت می‌کنه.  پس بیاید قبل از هرکاری ببینیم چقدر این داده‌ها تهی هستن.  برای این کار تعداد و درصد داده‌های تهی هر ستون رو بگید. 
(۴ امتیاز)

```python
empty_count = jobs.shape[0] - jobs.count()
empty_percent = (empty_count / jobs.shape[0]) * 100
empty_stats = pd.DataFrame({'Empty_Cells_Count': empty_count, 'Empty_Cells_Percent': empty_percent}, index=jobs.columns)
empty_stats
```

<div dir="rtl">
    همون‌طور که می‌بینید در بعضی از ستون‌ها بیش از ۵۰ درصد داده تهی هستن. ستون‌هایی که این ویژگی رو دارن از مجموعه داده‌ها حذف کنید.
    (۴ امتیاز)
</div>

```python
almost_filled_columns = empty_stats.query("Empty_Cells_Percent <= 50").index
jobs_with_filled_columns = jobs[almost_filled_columns]
jobs_with_filled_columns
```

<div dir="rtl">
در بعضی از ستون‌ها فقط یک داده‌ی تهی وجود داره. سطرهای مربوط به این داده‌های خالی رو از مجموعه‌ی داده‌ها حذف کنید و ابعاد مجموعه داده‌های جدید را نمایش بدید.
    (۴ امتیاز)
</div>

```python
columns_with_one_empty_cell = empty_stats.query("Empty_Cells_Count == 1").index
cleaned_jobs = jobs_with_filled_columns.dropna(axis=0, subset=columns_with_one_empty_cell)
cleaned_jobs.shape
```

<div dir="rtl">
بعضی اوقات خالی بودن یک خانه از جدول می‌تواند معنادار و حاوی اطلاعات باشد. مثلا فرض کنید در یک مجموعه داده فرضی مربوط به مشتریان یک فروشگاه و لیست خریدهایشان، ستونی مربوط به قیمت پرداخت شده باشد و در ذهن طراح این دیتاست،‌ خالی بودن این خانه به این معناست که مشتری هنوز مبلغی را پرداخت نکرده است. در این صورت عملا با Data Missing مواجه نیستیم بلکه با یک داده که با فرمت نامناسبی نمایش داده شده مواجه هستیم. در این شرایط بهترین راهکار recode and reformat است. مثلا میتوانیم یک ستون جدا ایجاد کنیم که مشخص می‌کند آیا مبلغ پرداخت شده یا نه. که مقدارهای Yes یا No می‌تواند بگیرد. و اگر جواب No بود،‌مبلغ پرداختی را برابر با ۰ بگذاریم و به این صورت خانه‌های خالی جدول را پر کنیم. 
در جدول مربوط به سوال ما (usjobs) چنین حالتی فقط در مورد ستون Easy Apply ممکن است رخ داده باشد. برخی سطرها مقدار True دارند و سایر سطرها مقدار NaN که احتمال دارد در واقع بیانگر False باشد. این احتمال را از آن جا می‌دهم که هیچ‌جا در تضاد با  True، مقدار False ذکر نشده است و ممکن است در واقع شرکت‌هایی که اپلای راحتی ندارند، از جواب به این سوال امتناع کرده باشند در حالی که جوابشان در واقع False بوده است. اما خب این تنها یک حدس است و نیاز به بررسی بیشتری دارد.

در سایر موارد، دو ایده کلی برای مواجهه با Data Missing مطرح می‌شود که عبارت است از Removing و Imputation که در ادامه به آن‌ها می‌پردازیم. Imputation سعی میکند به نحوری داده خالی را پر کند، در حالی که removing به نوعی خانه خالی را حذف می‌کند. در ابتدا به حالات کلی Data Missing کمی بپردازیم. ۳ حالت کلی می‌توان متصور شد:

۱. Missing at Random:
 به شرایطی اشاره دارد که داده گم شده به مقدار خود داده ربطی ندارد، اما متاثر از برخی داده‌های مشاهده‌شده دیگر است. به عبارتی مقدار گم‌شده یک سطر به بقیه سطرهای جدول مرتبط می‌شود.
۲. Missing Completely at Random:
در این حالت،‌ گم شدن داده کاملا تصادفی بوده و به مقدار محتمل داده و یا سایر مشاهدات ربطی ندارد.
۳. Missing not at Random:
گم شدن داده رندوم نبوده و به سایر متغیرها بستگی دارد. مثلا بانوان علاقه‌ای به ذکر سن یا وزن خود ندارند پس گم شدن این مقادیر به جنسیت ممکن است ربط داده شود. یا مثلا افرادی که درآمد بسیار بالایی دارند معمولا  این مقدار را ذکر نمی‌کنند.

اگر گم‌شدن داده ناشی از MAR یا MCAR باشد،‌ معمولا روش‌های Removing بهتر جواب می‌دهد. اما استفاده از این روش‌های برای MNAR به شدت گمراه‌کننده می‌تواند باشد. مثلا مثال درآمد افراد پولدار را که در چند خط بالاتر گفته شد در نظر بگیرید،‌ حذف این مشاهدات میانگین درآمد افراد محله را به شدت تغییر خواهد داد. پس شاید بهتر باشد از روش‌های Imputation استفاده کنیم.
دقت شود فهمیدن این موضوع که علت گم شدن داده کدام یک از حالت‌های گفته شده است،‌در برخی موارد بسیار چالش‌انگیز است.

انواع روش‌های Removing

Listwise Deletion:
داده‌ای که یک یا چند خانه خالی دارد را به کلی از مجموعه حذف می‌کنیم. این‌ کار را در رابطه با ۲ تا از سطرها در این تمرین انجام دادیم. یکی از آن‌ها مقدار Salary Estimate و دیگر مقدار Company Name را نداشت و برای همین هر دو سطر را حذف کردیم. دقت شود که این کار نباید در مقیاس زیاد انجام شود. مثلا در این تمرین از بین ۲۲۵۳ سطر،‌ ۲ سطر را حذف کردیم. در غیر این صورت داده‌ها را بایاس می‌کند.
این روش تقریبا رایج‌ترین روش موجود است.

Pairwise Deletion:
در این روش سطرهای ناقص (شامل داده گم‌شده) تنها در شرایطی در نظر گرفته نمی‌شوند (به نوعی حذف می‌شوند) که از آن مقدار حذف شده بخواهیم برداشتی یا تحلیل آماری بکنیم. مثلا در همین تمرین تا اینجای کار،‌ بسیاری از سطرها هستند که مقادیر Sector یا Founded یا … آن‌ها NaN است اما چون این متغیرها در تحلیل‌‌های ما هنوز وارد نشده‌اند، داده‌های مربوطه را حذف نکرده‌ایم.
مزیت این روش این است که مثلا در تحلیل Salary Estimate، داده‌های بیشتری در اختیار داریم و احتمالا نتیجه تحلیل‌هایمان درست‌تر باشد. اما عیب و به نوعی سختی این روش در این است که اگر در تحلیل‌های بعدی بخواهیم مثلا از متغیر Sector داده‌ها نتیجه‌ای بگیریم، باید حواسمان باشد که حدود ۳۵۰ تا از داده‌‌ها این مقدار را ندارند و چون در Salary Estimate آن‌ها را لحاظ کردیم،‌اگر بخواهیم مثلا رابطه‌ای بین Sector و Salary Estimate پیدا کنیم به مشکل برمیخوریم و مثلا شاید مجبور بشویم با استفاده از روش‌های imputation مقادیر گم‌شده را پر کنیم که خب مشکلات خودش را به همراه دارد.


Dropping Variables:
اگر یک ستون (متغیر) درصد بالایی (مثلا بالای 60%) مقدار گم‌شده داشته باشد، می‌توان به جای حذف کل سطر مربوط با آن داده، کلا آن ستون را drop (حذف) کنیم. اما دقت شود که این در صورتی است که آن ستون خیلی داده مهمی نباشد. اما خب اگر مهم باشد،‌ پس نباید بیشتر از مثلا 50 درصد مقدار گم‌شده داشته باشد! از این روش برای حذف ستون‌های Competitors و Easy Apply استفاده شد که یکی بالی ۷۵ درصد و دیگری بالای ۹۰ درصد داده گم شده داشتند.

انواع روش‌های Imputation
حال قصد داریم به نوعی خانه خالی مربوط به داده گم‌شده را پر کنیم. اما خوب است قبل از بیان جزئیات این نکته را در نظر بگیریم که به هر حال داده گم‌شده اتفاق ناگواری است و این روش‌ها تلاش می‌کنند خسارت را کمینه کنند اما همیشه آسیب ناشی از گم‌شدن داده باقی می‌ماند.

Last Observation Carried Forward & Next Observation Carried Backward:
از این دو روش برای داده‌های مربوط به یک سری زمانی استفاده می‌شود. در این روش به دنبال داده‌ای مشابه با داده گم‌شده رفته (که مثلا فقط در مقدار زمان یا step متفاوت باشند.) و مقدار گم‌شده داده در زمان t را برابر با مقدار داده مشابه در زمان t-1 یا t+1 قرار می‌دهیم. که خب مجموعه داده usjobs در این تمرین چنین ویژگی‌ای ندارد.

Mean, Median and Mod:
یک راه این است که میانگین یا میانه یا یک شاخص دیگر را از داده‌های اطراف داده گم‌شده(که مثلا با شاخص خاصی group شده‌اند)‌ یا از کل مجموعه داده‌ها به دست آورده و جایگزین مقدار گم‌شده کنیم.
این روش  حساسیت‌های زیادی دارد. مثلا اگر صرفا داده‌های گم‌شده را با میانگین کل داده‌ها جایگزین کنیم، واریانس داده‌ها کاهش یافته و شاید تحلیل به درد بخوری نتوان انجام داد. در مجموعه داده تمرین، Salary Estimate یکی از سطرها خالی است. یک راه برای پر کردن آن می‌تواند این باشد که میانگین حقوقی که شرکت مربوطه در سطرهای دیگر پرداخت می‌کند را جایگزین کنیم. البته اگر مثلا کلا ۱ شرکت دیگر با همان اسم وجود داشته باشد،‌احتمالا این کار مناسبی نباشد.

Linear Regression:
به روش‌های یادگیری ماشین داریم نزدیک می‌شویم. در این روش، بین یک متغیر و سایر پارامترها رابطه خطی فرض شده و با استفاده از یادگیری از طریق سایر سطرها که داده‌ گم‌شده ندارند، ضرایب مربوط به این رابطه خطی پیدا شده و به این ترتیب مقدار گم‌شده یک داده از روی سایر پارامترهایش تعیین می‌شود. اما خب این روش مشکلات بسیاری دارد. مثلا اینکه فرض وجود رابطه خطی خیلی وقت‌ها مدل مناسبی را نتیجه نمی‌دهد. یا اینکه یادگیری می‌تواند زمان‌بر و هزینه‌بر باشد.

K Nearest Neighbors: 
در روش KNN، یک ملاک برای فاصله در نظر گرفته می‌شود. از این ملاک فاصله استفاده می‌شود تا فاصله یک سطر که داده گم‌شده دارد با سایر سطرها پیدا شود. مثلا می‌توان ملاک فاصله را بر اساس اختلاف پارامتر Min Salary گذاشت. پس از اینکه فاصله‌ها پیدا شد،‌ میانگین پارامتر گم‌شده مورد نظر بین k کم‌فاصله ترین سطرها گرفته شده و جایگزین پارامتر گم‌شده در سطر مورد نظر می‌شود. برای مقادیر گسسته نیز می‌توان به جای میانگین از میانه استفاده کرد.
مشکل این روش این است که تعریف ملاک فاصله خوب بسیار دشوار است و خودش نیاز به انجام آزمایشات بسیاری دارد و در بیشتر موارد اصلا مدلسازی مناسبی نیست و جواب خیلی خاصی نمی‌دهد.
مشکل دیگر هم این است که بسیار روش زمان‌بری است.

</div>

<div dir="rtl">
تا اینجای کار فقط داشتیم داده‌هامون رو تمیز می‌کردیم. از اینجا به بعد می‌خوایم کمی روی اونها کار کنیم. یکی از 
فاکتورهای مهم توی کار کردن میزان حقوقه. پس بیاید و ستون تخمین حقوق رو به دو ستون تبدیل کنید که حداقل و حداکثر حقوق 
درخواستی در اون درج شده.
سپس ستون میانگین حقوق را از این دو ستون جدید به دست بیارید و در انتها پنج داده‌ی اول مجموعه داده رو نمایش بدید.
    (۸ امتیاز)
</div>

```python
def get_min_salary(salary_range):
    return salary_range.split("-")[0]


def get_max_salary(salary_range):
    return salary_range.split("-")[1].split(" ")[0]


def get_mean_salary(min, max):
    min = int(min[1:-1])
    max = int(max[1:-1])
    mean = (max + min) / 2
    return '$' + str(mean) + 'K'


min_salary_series = cleaned_jobs["Salary Estimate"].map(get_min_salary)
max_salary_series = cleaned_jobs["Salary Estimate"].map(get_max_salary)

jobs_with_min_salary = cleaned_jobs.assign(Min_Salary=min_salary_series)
jobs_with_min_max_salary = jobs_with_min_salary.assign(Max_Salary=max_salary_series)

min_max_salary = jobs_with_min_max_salary[["Min_Salary", "Max_Salary"]]
mean_salary_series = min_max_salary.apply(lambda row: get_mean_salary(row['Min_Salary'], row['Max_Salary']), axis=1)

jobs_with_salary_details = jobs_with_min_max_salary.assign(Mean_Salary=mean_salary_series)
jobs_with_salary_details = jobs_with_salary_details.drop('Salary Estimate', 1)
jobs_with_salary_details.head(5)
```

```python
def get_revenue_mean(range):
    range1 = range
    if type(range) != str:
        return -1
    range = range.split(" ")
    if range[0] == "Unknown":
        return -1
    elif range[0] == "Less":
        if range[3] == "billion":
            k = 1000
        else:
            k = 1
        mean = range[2]
        return int(mean[1:]) * k
    elif len(range) == 3:
        if range[1] == "billion":
            k = 1000
        else:
            k = 1
        mean = range[0][0:-1]
        return int(mean[1:]) * k
    elif len(range) == 6:
        min = range[0]
        max = range[3]
        return ((int(min[1:]) + int(max[1:]) * 1000) / 2 )
    else:
        min = range[0]
        max = range[2]
        if range[3] == "billion":
            k = 1000
        else:
            k = 1
        try:
            return ((int(min[1:]) + int(max[1:])) / 2 ) * k
        except:
            print(range1)

companies["Revenue"] = jobs3.groupby("Company Name").first()["Revenue"]
companies["Average_Revenue(million USD)"] = companies["Revenue"].map(get_revenue_mean)
# revenue_median = companies.query("Cleaned_Revenue != -1")['Cleaned_Revenue'].median()
companies_with_revenue = companies[companies["Average_Revenue(million USD)"] != -1]

fig,ax = plt.subplots(figsize=(15,6))
sns.boxplot(x="Average_Revenue(million USD)", y="Average_Salary(thousend USD)", data=companies_with_revenue, ax=ax);
```

```python
def get_size_mean(range):
    range1 = range
    if type(range) != str:
        return -1
    range = range.split(" ")
    if range[0] == "Unknown":
        return -1
    elif len(range) == 2:
        mean = range[0][0:-1]
        return int(mean)
    else:
        min = range[0]
        max = range[2]
        try:
            return (int(min) + int(max)) // 2
        except:
            print(range1)


companies["Size"] = jobs3.groupby("Company Name").first()["Size"]
companies["Size(persons)"] = companies["Size"].map(get_size_mean)
companies_with_size = companies[companies["Size(persons)"] != -1]

fig,ax = plt.subplots(figsize=(15,6))
sns.boxplot(x="Size(persons)", y="Average_Salary(thousend USD)", data=companies_with_size, ax=ax);
```
